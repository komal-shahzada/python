```python
def bresenham(x0, y0, x1, y1):
    dx = x1 - x0
    dy = y1 - y0

    # Ensure dx is non-negative for the loop to work as intended
    # This implementation assumes x0 <= x1. If not, swap points.
    if dx < 0:
        x0, x1 = x1, x0
        y0, y1 = y1, y0
        dx = x1 - x0
        dy = y1 - y0

    # Handle cases where dy is negative, for consistency with p calculation
    # This version of Bresenham is typically for |slope| <= 1. Let's assume that for now.
    # For a general algorithm, a more robust implementation handling all octants is needed.

    p = 2*dy - dx
    x = x0
    y = y0

    points = []
    while x <= x1:
        points.append((x, y))
        x += 1
        if p < 0:
            p += 2*dy
        else:
            y += 1
            p += 2*dy - 2*dx
    return points

# Example usage:
# points = bresenham(0, 0, 8, 4)
# print(points)
```

```python
import matplotlib.pyplot as plt

# Get the points from the bresenham function
points = bresenham(0, 0, 8, 4)

# Separate x and y coordinates for plotting
x_coords = [p[0] for p in points]
y_coords = [p[1] for p in points]

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x_coords, y_coords, marker='o', linestyle='-', color='b')
plt.title('Bresenham\'s Line Algorithm Visualization (0,0) to (8,4)')
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.grid(True)
plt.xticks(range(min(x_coords), max(x_coords) + 1))
plt.yticks(range(min(y_coords), max(y_coords) + 1))
plt.show()
```
