```python
#sets
fruits = {"apple", "banana", "cherry"}
```

```python
# for loop
fruits = {"apple", "banana", "cherry"}

for fruit in fruits:
    print(fruit)
```

```python
# while loop in python
fruits = {"apple", "banana", "cherry"}
fruit_list = list(fruits)

i = 0
while i < len(fruit_list):
    print(fruit_list[i])
    i += 1
```

```python
#loop with condition
numbers = {1, 2, 3, 4, 5}

for n in numbers:
    if n % 2 == 0:
        print(f"{n} is even")
```
