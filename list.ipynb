```python
# Create a list
students = ["Atif", "Ali", "Sara"]

# Add a new student
students.append("Ahmed")

# Replace a student name
students[1] = "Hassan"

# Remove one
students.remove("Sara")

# Print all students
for s in students:
    print(s)

print("Total students:", len(students))
```

```python
# Create an empty list to store numbers
numbers = []          # This initializes an empty list named 'numbers' where we will store user inputs

# Take 5 numbers as input from the user
for i in range(5):    # A loop that will run 5 times (from 0 to 4)
    # Ask the user to enter a number
    num = float(input(f"Enter number {i+1}: "))
    # input() takes user input as text (string)
    # float() converts it to a decimal number (so we can calculate average later)
    # f"Enter number {i+1}" shows the message like "Enter number 1", "Enter number 2", etc.

    numbers.append(num)   # Add the entered number to the 'numbers' list

# Print the full list entered by the user
print("\nNumbers you entered:", numbers)
# '\n' means a new line before printing the list (for clean output)

# Find the maximum, minimum, and average
maximum = max(numbers)      # max() finds the largest number in the list
minimum = min(numbers)      # min() finds the smallest number in the list
average = sum(numbers) / len(numbers)
# sum(numbers) adds all numbers together
# len(numbers) gives total count of items (here itâ€™s 5)
# dividing sum by length gives the average

# Display the results
print("Maximum number:", maximum)     # prints the largest number
print("Minimum number:", minimum)     # prints the smallest number
print("Average number:", average)     # prints the average value
```
