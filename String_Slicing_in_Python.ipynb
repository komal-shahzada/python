```python
a='Hello word'
print(a[3::5])
```

```python
a = "Hello"
print(a[::2])
```

```python
text = "Programming"
print(text[0:6])
# Output: Progra
# Explanation: Starts at index 0, ends at index 6 (exclusive).
```

```python
text = "Programming"
print(text[:6])
# Output: Progra
# Explanation: Default start is 0.
```

```python
text = "Programming"
print(text[6:])
# Output: mming
# Explanation: From index 6 to the end of the string.
```

```python
text = "Python"
print(text[:])
# Output: Python
# Explanation: [:] returns the entire string (start = 0, end = end of string).
```

```python
text = "ABCDEFGHIJ"
print(text[0:9:2])
# Output: ACEGI
# Explanation: Step = 2, so every 2nd character is picked.
```

```python
text = "HelloWorld"
print(text[-5:])
# Output: World
# Explanation: -5 means 5th character from the end.
```

```python
text = "Python"
print(text[::-1])
# Output: nohtyP
# Explanation: Step = -1 reverses the string.
```

```python
text = "PythonProgramming"
print(text[10:3:-1])
# Output: gnirorP
# Explanation: Starts at index 10, moves backward until index 3.
```

```python
text = "DataScience"
print(text[::3])
# Output: DtiC
# Explanation: Every 3rd character is taken.
```

```python
text = "MachineLearning"
print(text[-1:-10:-2])
# Output: giLh
# Explanation: Starts from last character, goes backwards, step = -2.
```
